```python
# ===============================
# TASK 1: Create Dataset & Model
# ===============================

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 1: Create synthetic dataset
np.random.seed(42)

n = 60

area_sqft = np.random.randint(500, 3000, n)
num_bedrooms = np.random.randint(1, 5, n)
age_years = np.random.randint(0, 30, n)

# Target variable (price in lakhs)
price_lakhs = (
    area_sqft * 0.05 +
    num_bedrooms * 10 -
    age_years * 0.3 +
    np.random.normal(0, 5, n)
)

# Step 2: Create DataFrame
df = pd.DataFrame({
    'area_sqft': area_sqft,
    'num_bedrooms': num_bedrooms,
    'age_years': age_years,
    'price_lakhs': price_lakhs
})

print("First 5 rows of dataset:")
print(df.head())

# Step 3: Features & Target
X = df[['area_sqft', 'num_bedrooms', 'age_years']]
y = df['price_lakhs']

# Step 4: Train model
model = LinearRegression()
model.fit(X, y)

# Step 5: Print intercept & coefficients
print("\nIntercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Step 6: Predictions
y_pred = model.predict(X)

# Step 7: First 5 actual vs predicted
result = pd.DataFrame({
    'Actual': y.head(),
    'Predicted': y_pred[:5]
})

print("\nFirst 5 Actual vs Predicted:")
print(result)


# ===============================
# TASK 2: Evaluation Metrics
# ===============================

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print("\nModel Evaluation:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)

# Explanation:
# MAE shows the average absolute difference between actual and predicted values.
# RMSE gives higher weight to large errors, so it penalizes big mistakes more.
# R2 score tells how well the model explains the variance in data (closer to 1 is better).


# ===============================
# TASK 3: Residuals & Histogram
# ===============================

import matplotlib.pyplot as plt

# Residuals
residuals = y - y_pred

# Plot histogram
plt.hist(residuals)
plt.title("Residuals Distribution")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()

# Explanation:
# Residual = Actual - Predicted value.
# If the histogram is roughly bell-shaped, the model is performing well.
# If it's skewed or irregular, the model may not be capturing patterns properly.
```
